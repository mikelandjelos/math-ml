**Задатак 5.** Ретко поседнуте матрице представљају оне које имају велики садржај нула елемената.

**Дефиниција 1.** (Вилкинсон): Матрица може да се сматра ретко поседнутом када се ненула елементи налазе или концентрисано у уском појасу око дијагонале, или су распоређени на мање систематичан начин. Матрицу ћемо називати густом ако проценат нула-елемената или њихова расподела није таква да је економично искористити њихово присуство.

**Дефиниција 2.** Матрица $A\in\mathcal{M}_{m\times n}$ је ретко поседнута уколико је број ненула елемената $\mathcal{O}(m+n).$

У наредној кодној ћелији је дефинисана класа `RetkaMatrica` која имплементира тзв. **COO (Coordinate)** формат. Она има следеће карактеристике:
- `m`: број врста.
- `n`: број колона.
- `data`: NumPy низ димензије $k\times3$ који садржи податке о свим ненула елементима у облику `[vrednost, i, j]`.
- `k`: укупан број ненула елемената.
- `vrste` и `kolone`: помоћни низови индекса врста и колона које садрже податке.

Приликом иницијализације, класа аутоматски сортира елементе лексикографски (по врстама, па по колонама) и сабира вредности ако се појаве дупликати на истим координатама.

```python
class RetkaMatrica:
    def __init__(self, m, n, entries=None):
        """
        m: broj vrsta
        n: broj kolona
        entries: numpy niz dimenzije k x 3, gde je svaki red oblika [vrednost, i, j]
        """
        self.m = m
        self.n = n
        if entries is None:
            self.data = np.zeros((0, 3))
        else:
            self.data = np.array(entries, dtype=float)
            self._ukloni_duplikate()
            
    def _ukloni_duplikate(self):
        """
        Лексикографски сортира по (i, j) и сабира све вредности на истим позицијама.
        Задржава само по један елемент по индексу (i, j), уз збир као вредност.
        """
        if len(self.data) == 0:
            self.k = 0
            self.vrste = np.array([])
            self.kolone = np.array([])
            return

        i = self.data[:, 1].astype(int)
        j = self.data[:, 2].astype(int)

        # Лексикографско сортирање (прво по i, па по j)
        # np.lexsort користи последњи аргумент као примарни кључ
        sort_idx = np.lexsort((j, i))
        self.data = self.data[sort_idx]

        # Креирање јединствених 1D кључева за индексе (flattening)
        kljucevi = self.data[:, 1].astype(int) * self.n + self.data[:, 2].astype(int)
        
        # Проналажење граница група истих кључева
        _, start_idx = np.unique(kljucevi, return_index=True)

        # Сабирање вредности по групама (користимо reduceat за брзину)
        vrednosti = self.data[:, 0]
        zbir = np.add.reduceat(vrednosti, start_idx)

        # Реконструисање i и j из кључева (за позиције где су сабране вредности)
        # Узимамо само индексе на почецима група јер су унутар групе исти
        rekonstruisani_kljucevi = kljucevi[start_idx]
        novi_i = rekonstruisani_kljucevi // self.n
        novi_j = rekonstruisani_kljucevi % self.n

        # Нови садржај data
        self.data = np.column_stack((zbir, novi_i, novi_j))

        # Ажурирање атрибута
        self.vrste = np.unique(novi_i)
        self.kolone = np.unique(novi_j)
        self.k = len(self.data)

    def __repr__(self):
        return f"RetkaMatrica({self.m}x{self.n}) sa {self.k} nenula elemenata \n{self.data}"

    def puna_matrica(self):
        """Vraća gustu matricu (numpy array) za potrebe provere."""
        dense = np.zeros((self.m, self.n))
        if self.k > 0:
            dense[self.data[:,1].astype(int), self.data[:,2].astype(int)] = self.data[:,0]
        return dense
```
