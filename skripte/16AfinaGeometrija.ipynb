##  Идеја иза афиних простора

Векторски простори и потпростори довољни су за описивање скупа решења хомогеног система линеарних једначина. За нехомогене системе користимо надоградљу векторских простора: афине просторе и геометрију. Разлика настаје третирањем вектора као вектора положаја тачака. Додавањем појма тачке додајемо нове могућности векторима. Модел који користимо је аналитичка геометрија и $2D$ и $3D.$ Аналитичка геометрија поред појмова вектора и базе векторског простора користи појам тачке и координатног система.

Сваки вектор векторског простора са почетком у координатном почетку $O$ војим крајем одређује јединствену тачку. Због тога вршимо поистовећивање вектора са вектором положаја њихових крајњих тачака. Такве векторе зовемо везаним. Слободни вектори су усмерене дужи између две тачке простора. Док радимо исклјучиво са везаним векторима, моделирамо (скицирамо) векторски простор. Када уведемо слободне векторе и тачке, тада моделирамо афине просторе. 

Додавањем тачака векторском простору ширимо полазни скуп објеката, самим тим и операција над њима.  Векторски простор $V=\mathbb{R}^n$ проширен скупом тачака $\mathcal{V}$ и операцијом тачка$+$вектор (транслација) представља афини простор. Тачке ћемо означавати великим штампаним словима, а векторе и даље малим писаним словима латинице. Сваки вектор задат листом од $n$ бројева $\begin{bmatrix} x_0&x_1&\dots&x_{n-1}\end{bmatrix}^T,$ посматраћемо и као тачку афиног простора $\mathbb{R}^n.$ Дакле, тачка $T$ и њен вектор положаја $r_T=\overrightarrow{OT}$ су исти објекти. Скалари су и даље означени словима грчког алфабета. Особине операција у афином простору су природна последица операција над векторима:
$$\begin{array}{c} 
T+\theta=T,\quad (T+u)+v=T+(u+v),\\
\lambda T=\lambda\,\overrightarrow{OT}=\lambda\,r_T.
\end{array}$$
Операција транслације за фиксирани вектор $v$ је бијекција на скупу тачака,
$$\mathcal{T}_v:\ \mathcal{V}{\mapsto}\mathcal{V},\qquad \mathcal{T}_v(T)=T+v.$$
Због тога за произвољне две тачке $A$ и $B$ постоји јединствен вектор $v$ такав да је $B=A+v.$ Вектор $v=\overrightarrow{AB}$ зовемо слободан вектор.  Координате слободног вектора $\overrightarrow{AB}$ могу се одредити као разлика координата вектора положаја крајње и почетне тачке.

## Афини потпростори

**Дефиниција 1.**  Нека је $P\subseteq\mathbb{R}^n$ векторски потпростор и $T\in\mathbb{R}^n$ произвољна тачка, односно њен вектор положаја $r_T.$ Транслацију
$$L=T+P=\{T+v\ \mid\ v\in P\}=\{r_T+v\ \mid\ v\in P\}$$
зовемо афини потпростор простора $\mathbb{R}^n.$ Димензија афиног потпростора $L$ је димензија векторског потпростора $P.$

Веома важно својство афиног потпростора $L$ је: добијамо га транслацијом векторског простора $P$ у тачку $T,$ тј. за неки вектор полажаја $r_T.$ Векторски потпростор $P$ добијамо негативном транслацијом $L-r=L+(-r),$ или $r-L,$ афиног потпростора $L$ за произвољан вектор положај тачке $r\in L.$ 
Потпростор $P$ афиног простора $L=T+P=r_T+P$ још зовемо и директриса афиног потпростора $L.$ Осим тога, избор тачке транслације $T$ није битан, докле год је $T\in L.$ 

Афини потпростор је векторски (или линеаран) потпростор акко садржи нула-вектор $\theta.$
Аналитичка геометрија са тачком, правом и равни је модел геометрије афиних простора. Тачка је тако афини потпростор димензије нула, а ако је координатни почетак, онда је та тачка и тривијалан векторски потпростор $\{\theta\}.$ Било која права је афини потпростор димензије 1. Праве које пролазе кроз координатни почетак су поред тога и векторски потпростори. Свака раван је афини потпростор димензије $2$ и кодимензије $1,$ а раван кроз координатни почетак је и векторски потпростор са истим тим димензијама. Настављајући кроз просторе већих димензија на сличан начин замишљамо афине и векторске потпросторе.

Афини потпростор кодимензије $1$ (димензије $n-1$) зове се афина хиперраван. 

**Пример 1.** Једначина праве кроз две тачке $A\equiv r_A$ и $B\equiv r_B,$ са вектором положаја текуће тачке $X\equiv r,$ гласи: 
\begin{align}
		  &   &&\overrightarrow{AX}=\lambda\,\overrightarrow{AB}\ ,&&
		\lambda\in\mathcal{L},
  &\Longleftrightarrow&&	X-A=\lambda\,\overrightarrow{AB}\ ,
&&		\lambda\in\mathbb{R},\\
  &\Longleftrightarrow&&	X=A+\lambda\,\overrightarrow{AB}\ ,
&&		\lambda\in\mathbb{R}
&\Longleftrightarrow&&	r=r_A+\lambda(r_B-r_A)\ ,
&&		\lambda\in\mathbb{R}.
\end{align}
Тачка $A$ у изразу $X=A+\lambda\overrightarrow{AB}$ представља базну тачку праве јер је $X=A$ за $\lambda=0.$  За које $\lambda$ je $X=B?$ 

$$r\in r_A+\mathcal{L}(v),\quad v=r_B-r_A=\overrightarrow{AB}.$$

Вектор $\overrightarrow{AB}$ дефинише правац простирања ове праве, тј. директрису $\mathcal{L}(\overrightarrow{AB}).$ Скаларом $\lambda$ описује се колико пута од тачке $A$ треба надовезати вектор $\overrightarrow{AB}$ не би ли се дошло до тачке $X.$ Знак броја $\lambda$ одређује смер надовезивања. За $\lambda>0$ описују се тачке са полуправе од $A$ у смеру ка $B.$ Када је $\lambda<0$ описују се тачке полуправе која почиње у $A$ и не садржи $B.$  

Приметимо да се све тачке $X=A+\lambda\overrightarrow{AB}$ могу да напишу и у облику 
$$
X=(1-\lambda)A+\lambda B\quad\Longleftrightarrow\quad
r=(1-\lambda)r_A+\lambda\, r_B,\quad \lambda\in\mathbb{R}.
$$
Димензионалност афиног простора тада можемо да пратимо кроз број слободних параметара у линеарној комбинацији тачака.
За праву је потребан један параметар $\lambda\in\mathbb{R},$ тиме још једном потврђујемо да се ради о једнодимензионалном афином потпростору.

Нарена теорема потврђује геометријска својства афиних потпростора које имају тачка, права и раван у $3D.$

**Теорема 1.**
     Нека је $L\subseteq\mathbb{R}^n$ подскуп афиног простора. $L$ је афини потпростор акко за произвољне две тачке овог скупа $A,B\in L$ важи да афина права која пролази кроз $A$ и $B$ цела лежи у $L.$ 

Примећујемо аналогију у тестовима потпростора:
>$P$ је ралан векторски потпростор акко
$$
(\forall v,u\in P)(\forall\lambda,\mu\in\mathbb{R})(\lambda v+\mu u\in P).$$

>$L$ је афини потпростор акко
$$
(\forall A,B\in L)(\forall\lambda\in\mathbb{R})\big(\lambda A+(1-\lambda) B\in L\big).
$$


**Дефиниција 2.**
  Линеарне комбинације код којих је збир коефицијената једнак $1$ зовемо афине комбинације.  

Дајемо без доказа још неке веома важне карактеризације афиних потпростора.

**Теорема 2.** Најмањи афини потпростор који садржи дати скуп тачака $T_0,T_1,\dots,T_n$ је скуп свих афиних комбинација ових тачака. Означавамо га са \
$${\rm aff}(T_0,T_1,\dots,T_n)=\big\{\lambda_0T_0+\lambda_1T_1+\dots+\lambda_nT_n\ |\ \lambda_0+\lambda_1+\dots+\lambda_n=1\big\}.$$

**Теорема 3.**
     Скуп $L\subseteq\mathbb{R}^n$ је афини потпростор акко је скуп решења неког система линеарних једначина.

## Линеарне комбинације и омотачи

Множење вектора $v$ скаларом $\lambda\in(0,1]$ даје вектор $u=\lambda\, v$ истог правца и смера али не веће дужине у односу на вектор $v,$ 
\begin{align}
	    \cos\angle(v,\lambda v)&=\dfrac{v\cdot(\lambda v)}{\|v\|\,\|\lambda v\|}=\dfrac{\lambda(v\cdot v)}{|\lambda|\,\|v\|^2}\stackrel{\lambda\geq0}{=}\dfrac{\lambda\|v\|^2}{\lambda\|v\|^2}=1,\\[5pt]
\|\lambda v\|&=|\lambda|\,\|v\|\stackrel{\lambda\geq0}{=}\lambda\|v\|\stackrel{\lambda\leq1}{\leq}\|v\|.
\end{align}
Ова особина, као и неједнакости троугла
$$\big|\|v\|-\|u\|\big|\leq\|v+u\|\leq\|v\|+\|u\|,$$ 
користе се за описивање унутрашњости и спољашњости скупова вектора. У томе важну улогу имају специјалне линеарне комбинације.

Одсечак афине праве са кајњим тачкама $T_0,T_1\in\mathbb{R}^n$ зваћемо дуж или интервал и обележаваћемо по угледу на сегменте на реалној правој,
$$[T_0,T_1]=\big\{T\in\mathbb{R}^n\ |\ T=(1-\lambda)T_0+\lambda T_1,\ \lambda\in[0,1]\big\}.$$

**Дефиниција 3.** Скуп тачака $S$ афиног простора $\mathbb{R}^n$ је конвексан уколико за за сваке две тачке скупа $S$ дуж која их спаја цела припада скупу $S,$ тј.		
$$\big(\forall T_0,T_1\in S\big)
\big([T_0,T_1]\subseteq S\big).$$

**Дефиниција 4.** Нека су $v_0,v_1,\dots,v_k\in\mathbb{R}^n$ дати вектори и $\alpha_0,\alpha_1,\dots,\alpha_k\in\mathbb{R}$ неки скалари. Израз 
$$\alpha_0v_0+\alpha_1v_1+\dots+\alpha_kv_k\tag{1}$$
је линеарна комбинација. 


За скуп $S\subseteq\mathbb{R}^n$ скуп свих (коначних) линеарних комбинација вектора из скупа $S$ је **линеал** или **линеарно затварање**  тог скупа 
$$\mathcal{L}(S)=\big\{\alpha_0v_0+\alpha_1v_1+\dots+\alpha_kv_k\ |\ k\in\mathbb{N}_0,\ v_i\in S,\ \alpha_i\in\mathbb{R}\big\}.$$
С обзиром да радимо у коначно-димензионалном векторском простору $\mathbb{R}^n$ довољно је познавање неког потпуног скупа вектора за генерисање линеала,
$$\mathcal{L}(v_0,v_1,\dots,v_k)=\big\{\alpha_0v_0+\alpha_1v_1+\dots+\alpha_kv_k\ |\ \alpha_0,\alpha_1,\dots,\alpha_k\in\mathbb{R}\big\}.$$

Када коефицијенти $\alpha_0,\alpha_1,\dots,\alpha_k\in\mathbb{R}$ линеарне комбинације $(1)$ узимају неке посебне вредности добијамо скупове вектора или тачака са посебним својствима:
- Aко су коефицијенти $\alpha_0,\alpha_1,\dots,\alpha_k\geq0$ линеарна комбинација $(1)$ је **конусна**. 
- Када скалари задовољавају једнакост $\alpha_0+\alpha_1+\dots+\alpha_k=1$ линеарна комбинација $(1)$ је **афина**. 
- Линеарна комбинација $(1)$ је **конвексна** ако је конусна и афина, тј. ако је $\alpha_0,\alpha_1,\dots,\alpha_k\geq0$ и $\alpha_0+\alpha_1+\dots+\alpha_k=1.$ 

Скуп свих конусних комбинација вектора скупа $S$ је **линеаран конус** или **сектор** или **конусни омотач** тог скупа вектора, у ознаци 
$${\rm cone}(S)=\big\{\alpha_0v_0+\alpha_1v_1+\dots+\alpha_kv_k\ |\ k\in\mathbb{N}_0,\ v_i\in S,\ \alpha_i\geq0\big\}.$$
Када је $S$ коначан скуп, $S=\{v_0,v_1,\dots,v_k\},$ конусни омотач ${\rm cone}(S)$ зовемо **полиедар**.

Свим афиним комбинацијама тачака скупа $S$ генерише се њихов **афини омотач** или **афино затварање**
$${\rm aff}(S)=\big\{\alpha_0v_0+\alpha_1v_1+\dots+\alpha_kv_k\ |\ k\in\mathbb{N}_0,\ v_i\in S,\ \alpha_0+\alpha_1+\dots+\alpha_k=1\big\}.$$

Скуп свих конвексних комбинација тачака скупа $S$ је **конвексни омотач** или **конвексно затварање** 
$${\rm conv}(S)=\left\{\alpha_0v_0+\alpha_1v_1+\dots+\alpha_kv_k\ |\ k\in\mathbb{N}_0,\ v_i\in S,\ \sum_{i=0}^k\alpha_i=1,\ \alpha_i\geq0\right\}.$$
Када је $S$ коначан скуп, $S=\{v_0,v_1,\dots,v_k\},$ конвексни омотач ${\rm conv}(S)$ зовемо **конвексни политоп**.

Наведене линеарне комбинације служе за детекцију битних скупова вектора и тачака.
- Скуп вектора је линеаран потпростор акко садржи све (коначне) линеарне комбинације својих елемената.
- Скуп тачака је афини потпростор акко садржи све (коначне) афине комбинације својих елемената.
- Скуп тачака је сектор акко садржи све (коначне) конусне комбинације својих елемената.
- Скуп тачака је конвексан акко садржи све (коначне) конвексне комбинације својих елемената.

За скупове вектора, односно тачака, затварања одређеног типа имају својство минималности у класи скупова истог типа. Другим речима, линеарно затварање скупа вектора је најмањи линеаран потпростор који садржи те векторе. Афино затварање је најмањи афини потпростор који садржи дате тачке, итд. Због тога, упознаћемо ближе својства истакнутих линеарних комбинација кроз графички приказ.

**Пример 2.** Креираћемо омотаче (линеал,сектор,афини омотач, конвексни омотач) једног вектора у 2D. Резултати линеарних комбинација биће приказани тачкама - крајњим тачкама резултујућих вектора. У те сврхе дефинишемо функцију која ће постављени задатак обавити над произвољним скупом вектора димензије $2.$ Улаз у функцију је матрица чије колоне представљају дати скуп вектора. С обзиром да не могу да се израчунају све могуће линеарне комбинације, узети у обзир да графички прикази дају коначну количину података о одговарајућим затварањима.

```python
def kombinacije2D(A):
    _,n=A.shape
    plt.figure(figsize=(8,8))

    plt.subplot(2,2,1)
    plt.title("Линеарне комбинације")
    k=35
    koef=np.linspace(-5,5,k).reshape(1,k)
    for i in range(n-1):
        pom=np.repeat(koef[-1],k)
        koef=np.block([[np.tile(koef,k)],[pom]])
    tacke=A@koef
    plt.plot(tacke[0],tacke[1],'o')
    plt.plot(A[0],A[1],'or')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.xlim([-7,7])
    plt.ylim([-7,7])


    plt.subplot(2,2,2)
    plt.title("Конусне комбинације")
    k=30
    koef=np.linspace(0,5,k).reshape(1,k)
    for i in range(n-1):
        pom=np.repeat(koef[-1],k)
        koef=np.block([[np.tile(koef,k)],[pom]])
    tacke=A@koef
    plt.plot(tacke[0],tacke[1],'o')
    plt.plot(A[0],A[1],'or')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.xlim([-7,7])
    plt.ylim([-7,7])

    plt.subplot(2,2,3)
    plt.title("Афине комбинације")
    k=25
    koef=np.linspace(-5,5,k).reshape(1,k)
    for i in range(n-1):
        pom=np.repeat(koef[-1],k)
        koef=np.block([[np.tile(koef,k)],[pom]])
    s=np.sum(koef,axis=0)
    koef=koef[:,s!=0]
    s=s[s!=0]
    koef=koef/s
    tacke=A@koef
    plt.plot(tacke[0],tacke[1],'o')
    plt.plot(A[0],A[1],'or')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.xlim([-7,7])
    plt.ylim([-7,7])

    plt.subplot(2,2,4)
    plt.title("Конвексне комбинације")
    k=10
    koef=np.linspace(0,1,k).reshape(1,k)
    for i in range(n-1):
        pom=np.repeat(koef[-1],k)
        koef=np.block([[np.tile(koef,k)],[pom]])
    s=np.sum(koef,axis=0)
    koef=koef[:,s!=0]
    s=s[s!=0]
    koef=koef/s
    tacke=A@koef
    plt.plot(tacke[0],tacke[1],'o')
    plt.plot(A[0],A[1],'or')
    plt.axhline(color='gray', zorder=-1)
    plt.axvline(color='gray', zorder=-1)
    plt.xlim([-7,7])
    plt.ylim([-7,7]);
```

Генеришемо сада случајни вектор дужине $2$ чије омотаче тражимо. Он је на графицима означен црвеном тачком. Резултати линеарних комбинација су приказани плавим тачкама. 

```python
n=1  #menjanjem broja vektora mozemo da vidimo njihove linearne kombinacije
A=randint(-3,4,[2,n])
print(A)
kombinacije2D(A)
```

Примећујемо да је конусна комбинација једног ненула вектора полуправа са почетком у координатном почетку и која садржи тачку са датим вектором положаја. Полуправе са почетком у координатном почетку у геометрији већих димензија зовемо зрак. Ако је зрак транслацијом измештен из координатног почетка (почетак му није координатни почетак) зовемо га афини зрак. 

Афине и конвексне комбинације једне тачке су та тачка.

**Задатак 3.**  Нека су $T_0,T_1,\dots,T_k\in\mathbb{R}^n$ тачке простора и $\lambda_0,\lambda_1,\dots,\lambda_k$ реални бројеви не-нула суме, $\lambda_0+\lambda_1+\dots+\lambda_k=\lambda\neq0.$ Доказати да постоји јединствена тачка $T$ таква да је $$\lambda_0\overrightarrow{TT_0}+\lambda_1\overrightarrow{TT_1}+\dots+
		\lambda_k\overrightarrow{TT_k}=\theta,$$
где је $\theta$ нула-вектор.

Доказати да важи 
		$$T=\dfrac{1}{\lambda}(\lambda_0T_0+\lambda_1 T_1+\dots+\lambda_k T_k),$$
тј. да је $T$ барицентар тачака $T_0,T_1,\dots,T_k$ пондерисаних са $\lambda_0,\lambda_1,\dots,\lambda_k.$ У случају када скаларе $\lambda_0,\lambda_1,\dots,\lambda_k$ тумачимо као масе материјалних тачака $T_0,T_1,\dots,T_k,$ тачку $T$ зовемо центар масе тог система.

**Решење:** 

Посматраћемо дати услов као једначину по непознатој тачки $X$ за коју важи 
$$\lambda_0\overrightarrow{XT_0}+\lambda_1\overrightarrow{XT_1}+\dots+
		\lambda_k\overrightarrow{XT_k}=\theta.$$
Тражено $X$ можемо да добијемо решавањем ове једначине.  

С обзиром да је $\overrightarrow{XT_i}=T_i-X,\ i=0,1,\dots,k,$ то је
\begin{align} \theta&=\lambda_0\overrightarrow{XT_0}+\lambda_1\overrightarrow{XT_1}+\dots+
		\lambda_k\overrightarrow{XT_k}\\
        &=\lambda_0(T_0-X)+\lambda_1(T_1-X)+\dots+\lambda_k(T_k-X)\\
        &=\lambda_0T_0+\lambda_1 T_1+\dots+\lambda_k T_k-(\lambda_0+\lambda_1+\dots+\lambda_k)X\\
        &=\lambda_0T_0+\lambda_1 T_1+\dots+\lambda_k T_k-\lambda X
\end{align}
Одатле је 
$$X=\dfrac{1}{\lambda}(\lambda_0T_0+\lambda_1 T_1+\dots+\lambda_k T_k)=T.$$
С обзиром на поступак решења, закључујемо да је тачка $T$ јединствено решење дате једначине.

**Задатак 4.**  Дата су темена троугла $A,B,C$ (својим векторима положаја) и познате су дужине $a,b,c$ наспрамних страна, $$a=BC,\quad b=AC,\quad c=AB.$$ Доказати да је центар уписане кружнице $S$ барицентар скупа темена са тежинама које су једнаке дужинама страница $a,b,c:$
	$$S=\frac{aA+bB+c\,C}{a+b+c}\,.$$
<div>
     
</div>

**Решење:**

Да бисмо показали да је $S$ центар уписане кружнице троугла $\Delta ABC$ довољно је да покажемо да се $S$ налази на све три симетрале углова овог троугла. Показаћемо да $S$ лежи на симетрали угла код темена $B,$ за остала темена доказ се изводи аналогно. Другим речима, с обзиром да је избор ознака темена био произвољан, доказано за $B$ важиће и за преостала два темена.
$$\angle SBC=\angle SBA\quad\Longleftrightarrow\quad\cos\angle SBC=\cos\angle SBA
\quad\Longleftrightarrow\quad\frac{\overrightarrow{BS}\cdot\overrightarrow{BC}}{\|\overrightarrow{BS}\|\,\|\overrightarrow{BC}\|}=\frac{\overrightarrow{BS}\cdot\overrightarrow{BA}}{\|\overrightarrow{BS}\|\,\|\overrightarrow{BA}\|}$$
$$\Longleftrightarrow\quad\frac{\overrightarrow{BS}\cdot\overrightarrow{BC}}{a}=\frac{\overrightarrow{BS}\cdot\overrightarrow{BA}}{c} $$

$$S=\frac{aA+bB+cC}{a+b+c}\quad\Longleftrightarrow\quad(a+b+c)S=aA+bB+cC\quad\Longleftrightarrow\quad
(a+b+c)S-(a+b+c)B=aA+bB+cC-(a+b+c)B$$
$$\Longleftrightarrow\quad(a+b+c)(S-B)=a(A-B)+c(C-B)\quad\Longleftrightarrow\quad
\overrightarrow{BS}=\frac{a}{a+b+c}\overrightarrow{BA}+\frac{c}{a+b+c}\overrightarrow{BC}
$$
Тада је
\begin{align}
\frac{\overrightarrow{BS}\cdot\overrightarrow{BC}}{a}&=\frac1a\left(\frac{a}{a+b+c}(\overrightarrow{BA}\cdot\overrightarrow{BC})+\frac{c}{a+b+c}(\overrightarrow{BC}\cdot\overrightarrow{BC})\right)
=\frac1a\left(\frac{a}{a+b+c}(\overrightarrow{BA}\cdot\overrightarrow{BC})+\frac{c}{a+b+c}a^2\right)\\
&=\frac{\overrightarrow{BA}\cdot\overrightarrow{BC}}{a+b+c}+\frac{ac}{a+b+c}\\[12pt]
\frac{\overrightarrow{BS}\cdot\overrightarrow{BA}}{c}&=\frac1c\left(\frac{a}{a+b+c}(\overrightarrow{BA}\cdot\overrightarrow{BA})+\frac{c}{a+b+c}(\overrightarrow{BC}\cdot\overrightarrow{BA})\right)
=\frac1c\left(\frac{c}{a+b+c}(\overrightarrow{BA}\cdot\overrightarrow{BC})+\frac{a}{a+b+c}c^2\right)\\
&=\frac{\overrightarrow{BA}\cdot\overrightarrow{BC}}{a+b+c}+\frac{ac}{a+b+c},
\end{align}
па заиста важи $\angle SBC=\angle SBA.$

**Дефиниција 5.** Центар или средина скупа вектора $v_0,v_1,\dots,v_k,$ је вектор у ознаци $c=c(v_0,v_1,\dots,v_k)$ дат линеарном комбинацијом:
$$c=\dfrac1{k+1}v_0+\dfrac1{k+1}v_1+\dots+\dfrac1{k+1}v_k
	=\dfrac{v_0+v_1+\dots+v_k}{k+1}.$$

Горња једнакост даје представљање вектора $c$ преко вектора $v_0,v_1,\dots,v_k.$
Центар скупа вектора $c$ је уопштење аритметичке средине бројева. Предтсавља вектор положаја тачке $T$ која је конвексна комбинација тачака $T_0,T_1,\dots,T_k$ са векторима положаја $v_0,v_1,\dots,v_k.$ У овој конвексној комбинацији свака тачка има исти "удео" или тежину, центар јe вектор положаја тачке $T$ која припада конвексном омотачу ${\rm conv}(T_0,T_1,\dots,T_k).$

Посебно својство центра $c=c(v_0,v_1,\dots,v_k)$ описано је изразом:
$$
\sum_{i=0}^k(v_i-c)=(v_0-c)+(v_1-c)+\dots+(v_k-c)=\theta
$$
$$\Longleftrightarrow\quad \sum_{i\in[k]}\overrightarrow{TT_i}
 =\overrightarrow{TT_0}+\overrightarrow{TT_1}+\dots+\overrightarrow{TT_k}=\theta$$
Дакле, вектори разлика $v_i-c=\overrightarrow{TT_i}$ се међусобно анулирају. Због тога се центар скупа вектора још назива и тежиште у одговарајућим приликама.

Центар скупа вектора $c=c(v_0,v_1,\dots,v_k)$  има још једно занимљиво својство. То је особина минималности растојања центра $c$ од скупа $\{v_0,v_1,\dots,v_k\}.$     

**Теорема 4.** Центар $c=c(v_0,v_1,\dots,v_k)$  скупа вектора $v_0,v_1,\dots,v_k$ задовољава једнакост
$$
     \sum_{i\in[k]}\|v_i-c\|^2=\min_{v\in \mathbb{R}^n}\sum_{i\in[k]}\|v_i-v\|^2
     .
$$    

**DOKAZ:** 

Доказаћемо неједнакост која је еквивалентна траженој једнакости:
$$\forall v\in \mathbb{R}^n,\qquad
\sum_{i\in[k]}\|v_i-v\|^2\geq\sum_{i\in[k]}\|v_i-c\|^2.
$$
\begin{align}
   \displaystyle\sum_{i\in[k]}\|v_i-v\|^2&=\displaystyle\sum_{i\in[k]}(v_i-v)\cdot(v_i-v)=
   \displaystyle\sum_{i\in[k]}(v_i-c+c-v)\cdot(v_i-c+c-v)\\
   &=\displaystyle\sum_{i\in[k]}\left(
   \|v_i-c\|^2+(v_i-c)\cdot(c-v)+(c-v)\cdot(v_i-c)+\|c-v\|^2\right)\\
   &=\displaystyle\sum_{i\in[k]}\left(
   \|v_i-c\|^2+2(v_i-c)\cdot(c-v)+\|c-v\|^2\right)\\  
   &=\displaystyle\sum_{i\in[k]}
   \|v_i-c\|^2+2\displaystyle\sum_{i\in[k]}\big((v_i-c)\cdot(c-v)\big)+\displaystyle\sum_{i\in[k]}\|c-v\|^2\\
   &=\displaystyle\sum_{i\in[k]}\|v_i-c\|^2+(k+1)\|c-v\|^2+
   2\left(\displaystyle\sum_{i\in[k]}(v_i-c)\!\right)\!\cdot\!(c-v)\\
   &{=}\displaystyle\sum_{i\in[k]}\|v_i-c\|^2+(k+1)\|c-v\|^2+2\theta\cdot(c-v)\\
   &=\displaystyle\sum_{i\in[k]}\|v_i-c\|^2+(k+1)\|c-v\|^2\geq \displaystyle\sum_{i\in[k]}\|v_i-c\|^2.
\end{align}
У последњем реду у неједнакости, једнакост се очигледно постиже акко $\|c-v\|^2=0,$ тј. када је $v=c.$ Тиме је тражена особина доказана.

**Задатак 1.** Одредити једначину симетралне хиперравни дужи чије су крајње тачке $T_0$ и $T_1.$

**Решење :**

Нека су $r_0$ и $r_1$ вектори положаја тачака $T_0$ и $T_1.$ Симетрална хиперраван сачињена је од свих тачака које су на једнаком растојању од $T_0$ и $T_1.$ Означимо са $r$ вектор положаја произвољне тачке са ове хиперравни. Поставимо услов једнакости растојања:

\begin{align}
&\|r-r_0\|=\|r-r_1\|\quad\Longleftrightarrow\quad \|r-r_0\|^2=\|r-r_1\|^2
\quad\Longleftrightarrow\quad (r-r_0)\cdot(r-r_0)=(r-r_1)\cdot(r-r_1)\\
\Longleftrightarrow\quad& \|r_0\|^2-2(r\cdot r_0)=\|r_1\|^2-2(r\cdot r_1)
\quad\Longleftrightarrow\quad 2\,r\cdot(r_1-r_0)=\|r_1\|^2-\|r_0\|^2
\end{align}
Последња једнакост представља једначину тражене хиперравни. Приметимо да важи
\begin{align}\|r_1\|^2-\|r_0\|^2&=r_1\cdot r_1-r_0\cdot r_0=r_1\cdot r_1-r_1\cdot r_0+r_1\cdot r_0-r_0\cdot r_0\\
&=r_1\cdot(r_1-r_0)+r_0\cdot(r_1-r_0)=(r_1+r_0)\cdot(r_1-r_0).
\end{align}
Због тога, једначина хиперравни може да се напише и у облику
$$(2r-r_1-r_0)\cdot(r_1-r_0)=0.$$
