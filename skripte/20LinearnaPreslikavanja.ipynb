## Увод

Улога линеарних пресликавања у векторским просторима је аналогна улози непрекидних функција у математичкој анализи. Граничне вредности и непрекидност су кључне у анализи функција, тако су линеарни оператори кључни у анализи векторских простора. Непрекидне функције користимо да испитујемо геометријске особине и просторне релације које остају непромењене под континуалним деформацијама величине и облика објеката. У векторским просторима проучавамо линеарне трансформације јер оне подржавају њихову унутрашњу структуру. 

Линеарна пресликавања су алат за уочавање веза између векторских простора и конструкцију нових потпростора на основу постојећих. 

**Дефиниција 1.**
    Нека су $V$ и $U$ два векторска простора над истим пољем скалара $\mathbb{K}.$ Функција $\mathcal{A}:V\mapsto U$ је **хомогена** (homogeneous) ако важи
$$\big(\forall v\in V\big)\big(\forall \alpha\in \mathbb{K}\big)\quad  \mathcal{A}(\alpha v)=\alpha\mathcal{A}(v).$$
Функција $\mathcal{A}:V\mapsto U$ је **адитивна** (additive) ако важи
$$\big(\forall v,u\in V\big)\quad  \mathcal{A}( v+ u)=\mathcal{A}(v)+\mathcal{A}(u).$$

**Дефиниција 2.** Нека су $V$ и $U$ два векторска простора над истим пољем скалара $\mathbb{K}.$ Функција $\mathcal{A}:V\mapsto U$ је **линеарно пресликавање**, линеарна трансформација или **хомоморфизам** (linear map, linear transformation, homomorphism) векторских простора, уколико је хомогена и адитивна, тј. ако важи
    $$\big(\forall v,u\in V\big)\big(\forall \alpha,\beta\in \mathbb{K}\big)\quad  \mathcal{A}(\alpha v+\beta u)=\alpha\mathcal{A}(v)+\beta\mathcal{A}(u).\tag{1}$$

Вишеструком применом особине $(1)$ линеарног пресликавања $\mathcal{A}:V\mapsto U$ видимо да за њега важи:
$$(\forall n\in\mathbb{N})\big(\forall v_i\in V,\ i\in(n)\big)\big(\forall \alpha_i\in \mathbb{K},\ i\in(n)\big)\quad \mathcal{A}\left(\sum_{i\in(n)}\alpha_iv_i\right)=\sum_{i\in(n)}\alpha_i\mathcal{A}(v_i).$$

Када се векторски простори домена и кодомена поклапају, тј. $U=V,$ линеарно пресликавање $\mathcal{A}:V\mapsto V$ зваћемо **линеаран оператор** (linear operator). 

Када је кодомен поље скалара $U=\mathbb{K}$ посматрано као једнодимензионални простор, линеарно пресликавање $\mathcal{A}:V\mapsto \mathbb{K}$ зовемо **линеарна функционела** (linear functional, linear form).

**Пример 1.** **Идентичко пресликавање** $\iota:x\mapsto x$ је линеаран оператор на било ком векторском простору. Означаваћемо га $\mathcal{I}.$
    

Ако је $\theta_U$ нула-вектор векторског простора $U,$ константно пресликавање **нула-функција**
    $$(\forall v\in V)\ v\mapsto \theta_U$$ је линеарно пресликавање. Означаваћемо га са $\mathcal{O}.$ За линеарно пресликавање кажемо да је **нетривијално** ако није нула-пресликавање.

***

**Пример 2.** Ако је $A\in\mathcal{M}_{m\times n}$ матрица и $v\in\mathbb{R}^n$ произвољан вектор, тада је изразом
    $$\mathcal{A}(v)=Av$$
дато линеарно пресликавање $\mathcal{A}:\mathbb{R}^n\to\mathbb{R}^m.$ 

У случају квадратне матрице $A\in\mathcal{M}_{n\times n}$ пресликавање $\mathcal{A}:\mathbb{R}^n\to\mathbb{R}^n,$ $\mathcal{A}(v)=Av$ је линеаран оператор. 

Уколико је $a^T=A\in\mathcal{M}_{1\times n}$ пресликавање $v\mapsto a^Tv$ је линеарна функционела.

***

**Пример 3.** Скуп непрекидних функција $\mathbb{C}[a,b],$ скуп $n-$пута или бесконачно пута диференцијабилних функција $\mathbb{C}^n[a,b]$ и $\mathbb{C}^{\infty}[a,b],$ као и простор интеграбилних функција на интервалу - сви ови скупови представљају векторске просторе.

На простору реалних функција једне реалне променљиве многи од стандардних алата анализе имају својство линеарних оператора,
\begin{align}
       &\displaystyle\lim_{x\to a}\big(\alpha f(x)+\beta g(x)\big)
    =\alpha\displaystyle\lim_{x\to a}f(x)+\beta\displaystyle\lim_{x\to a}g(x),\\
    & \big(\alpha f(x)+\beta g(x)\big)'=\alpha f'(x)+\beta g'(x),\\
    & \displaystyle\int\big(\alpha f(x)+\beta g(x)\big)dx=\alpha\int f(x)dx+\beta\int g(x)dx.
\end{align}
Наведене једнакости указују на линеарна пресликавања на простору функција. 
   
Одређени интеграл, с обзиром да као резултат даје реалну вредност, 
$$\int_a^bf(x)dx\in\mathbb{R}$$
представља функционелу на простору интеграбилних функција. 

***

Нека су $V$ и $U$ два векторска простора над истим пољем скалара $\mathbb{K}.$ Скуп свих линеарних пресликавања између векторских простора $V$ и $U$ означавамо $L(V,U).$ За вежбу је згодно показати да је $L(V,U)$ потпростор векторског простора $U^V,$ свих пресликавања из $V$ у $U.$ За скуп линеарних оператора уместо да пишемо $L(V,V)$ то краће бележимо $L(V).$ Слично, за линеарне функционеле користимо скраћеницу $L(V,\mathbb{K})\equiv V'$ или $V^{\ast}.$

**Пример 4.** Нека су $V$ и $U$ векторски простори и $(b)$ база векторског простора $V.$ Уколико је $f:(b)\to U$ произвољно пресликавање, његова линеарна екстензија $L_f:V\to U$ је пресликавање које конструишемо преко слика базе $(b).$ За произвољан вектор $v\in V$ познато је његово представљање у односу на базу $(b):\{b_0,b_1,\dots,b_k\},$
    $$v=\alpha_0b_0+\alpha_1b_1+\dots+\alpha_k b_k.$$
Слику $L_f(v)$ тада добијамо изразом  
$$L_f(v)=\alpha_0f(b_0)+\alpha_1f(b_1)+\dots+\alpha_kf(b_k).$$

Овако дефинисана функција је добро дефинисана с обзиром на јединственост координата сваког вектора у односу на фиксирану базу. Једноставно се показује да је линеарна екстензија линеарно пресликавање. Закључујемо веома важну особину линеарних пресликавања: 
>  Свако линеарно пресликавање је довољно познавати на сликама неке базе.

Специјално, за две различите базе векторског простора $(a)$ и $(b)$ промена базе нпр. са $(a)$ у $(b)$ индукује једно линеарно пресликавање кроз линеарну екстензију: 
за $v=\alpha_0a_0+\alpha_1a_1+\dots+\alpha_k a_k,$ је
    $$L_f(v)=\alpha_0f(a_0)+\alpha_1f(a_1)+\dots+\alpha_kf(a_k)
    =\alpha_0b_0+\alpha_1b_1+\dots+\alpha_kb_k.$$

***

**Последица 1.** Нека су $V$ и $U$ векторски простори над истим пољем склара $\mathbb{K}$ и $(a)_{i\in I}$ база у $V.$ За произвољан скуп вектора $(u)_{i\in I}\subset U$ постоји јединствено линеарно пресликавање $\mathcal{A}:V\to U$ којим се слика $\mathcal{A}(a_i)=u_i,\ i\in I.$

## Особине линеарних пресликавања

**Лема 1.**  Линеарним пресликавањем се нула-вектор слика у нула-вектор. Парови супротних вектора се сликају у међусобно супротне векторе.
$$\mathcal{A}:V\to U\qquad \Longrightarrow\qquad \begin{cases}
\mathcal{A}(\theta_V)=\theta_U,\\
\mathcal{A}(-v)=-\mathcal{A}(v).
\end{cases}
$$

**Теорема 1.** Линеарним пресликавањем се векторски простор слика у векторски простор. Пред слика (инверзна слика) простора линеарним пресликавањем је векторски простор.

$$\mathcal{A}:V\mapsto U\quad\mbox{ и }\quad P\subseteq V,\  S\subseteq U\qquad\mbox{ потпростори}$$
тада су
$$\mathcal{A}(P)=\big\{\mathcal{A}(v)\ |\ v\in P\big\}\quad\mbox{ и }\quad\mathcal{A}^{-1}(S)=\big\{v\in V\ |\ \mathcal{A}(v)\in S\big\}\qquad\mbox{ потпростори}.$$

Сваки векторски простор има два тривијална потпростора, самог себе и нула-вектор $\{\theta\}.$ 
Теорема 1 је повод да скуп слика домена $\mathcal{A}(V)$ линеарног пресликавања носи име **простор слика**. Означавамо га још и $\mathcal{R}(\mathcal{A}),$ 
$$
\mathcal{R}(\mathcal{A})\equiv \mathcal{A}(V)=
\big\{\mathcal{A}(v)\ |\ v\in V\big\}.
$$

Ако за простор слика важи $\mathcal{A}(V)=U,$ линеарно пресликавање је "на" (onto, surjective), и зовемо га још и **епиморфизам** (epimorphism) векторских простора $V$ и $U.$ 

Према теореми 1 је и $\mathcal{A}^{-1}(\{\theta_U\})$ векторски потпростор. Зовемо га **језгро** пресликавања (null-space). Уобичајена ознака је:
$$
\mathcal{N}(\mathcal{A})=\mathcal{A}^{-1}(\{\theta_U\})=\big\{v\in V\ |\ \mathcal{A}(v)=\theta_U\big\}.
$$

Димензије ових потпростора су редом **ранг** и **дефект** линеарног пресликавања.
$${\rm rang}(\mathcal{A})=\dim\big(\mathcal{R}(\mathcal{A})\big),\qquad 
{\rm def}(\mathcal{A})=\dim\big(\mathcal{N}(\mathcal{A})\big).$$ 
Просторе $\mathcal{R}(\mathcal{A})$ и $\mathcal{N}(\mathcal{A})$ једним именом зовемо **фундаментални простори** линеарног пресликавања.

**Теорема 2.** Нека је дато линеарно пресликавање $\mathcal{A}:V\mapsto U.$
- $\mathcal{A}$ је сурјекција (**епиморфизам**) акко $\mathcal{A}$ слика потпун скуп вектора у потпун скуп. 
- $\mathcal{A}$ је "1-1", тј. инјекција (**мономорфизам**) акко  $\mathcal{A}$ слика линеарно независан скуп вектора у линеарно независан скуп. 
- $\mathcal{A}:V\mapsto U$ је "1-1"\ акко је $\mathcal{N}(\mathcal{A})=\{\theta_V\}.$

**Дефиниција 2.** Линеарно пресликавање векторских простора које је истовремено и бијекција назива се **изоморфизам**. Векторски простори су изоморфни уколико постоји изоморфизам између њих. Изоморфизам векторског простора на самог себе је **аутоморфизам**.

**Последица 2.** 
- Инверзно пресликавање изоморфизма је изоморфизам.
- Линеарно пресликавање је изоморфизам векторских простора акко слика базу у базу.
- Два векторска простора су изоморфна акко имају једнаке димензије.

**Пример 5.** Простор полинома $\mathbb{P}_n[t]$ и простор $\mathbb{R}^{n+1}$ су изоморфни. Један од могућих изоморфизама је линеарна екстензија пресликавања
    $$t^i\mapsto e_i,$$
где су $e_i$ вектори канонске базе простора $\mathbb{R}^{n+1}.$ Овим пресликавањем се полином слика управо у вектор коефицијената у стандардном облику,
    $$a_0+a_1t+\dots+a_nt^n\mapsto \begin{bmatrix} a_0&a_1&\dots&a_n\end{bmatrix}^T.$$

**За вежбу :** Пронаћи изоморфизам између векторских простора $\mathcal{M}_{m\times n}$ и $\mathbb{R}^{mn}.$

## Матрица линеарног пресликавања

**Дефиниција 3.** Нека је $(a):\{a_0,a_1,\dots,a_{n-1}\}$ база $n-$димензионалног векторског простора $V,$ и нека је $(b):\ \{b_0,b_1,\dots,b_{m-1}\}$ база $m-$димензионалног векторског простора $U$ над истим пољем скалара $\mathbb{K}.$ Нека је линеарно пресликавање између ових простора је $\mathcal{A}:V\to U.$ Координате слика $\mathcal{A}(a_k),\ k\in(n),$ изражене у односу на векторе базе $(b)$ смештене у колоне формирају матрицу $A\in\mathcal{M}_{m\times n}.$ Матрица $A$ назива се **матрица линеарног пресликавања** $\mathcal{A}$ у односу на базе $(a)$ и $(b),$
$$A=A\left(\mathcal{A},(a),(b)\right)=\left[\begin{array}{cccc}
\begin{bmatrix} \phantom{a}\\\mathcal{A}a_0\\ \phantom{a}\end{bmatrix}_{(b)}&
\begin{bmatrix} \phantom{a}\\\mathcal{A}a_1\\ \phantom{a}\end{bmatrix}_{(b)}&\dots&
\begin{bmatrix} \phantom{a}\\\mathcal{A}a_{n-1}\\ \phantom{a}\end{bmatrix}_{(b)}\end{array}\right].
$$

**Последица 3.** Матрица $A\in\mathcal{M}_{n\times n}$ линеарног оператора $\mathcal{A}$ је регуларна акко је оператор  $\mathcal{A}$ изоморфизам.

Када је линеарно пресликавање $\mathcal{A}:V\to U$ дато својом матрицом $A\in\mathcal{M}_{m\times n}$ у односу на базе $(a)$ и $(b),$ 
тада се израчунавање слика овим пресликавањем своди на координатне операције. За вектор $v\in V$ дат колоном координата $v_{(a)}$ 
$$v_{(a)}=x_0a_0+x_1a_1+\dots+x_{n-1}a_{n-1}\equiv\begin{bmatrix} x_0\\x_1\\\vdots\\x_{n-1}\end{bmatrix}_{(a)}$$
слика пресликавањем $\mathcal{A}$ имаће координате у бази $(b)$ дате изразом $Av_{(a)}.$
\begin{align}
(\mathcal{A}(v))_{(b)}&=Av_{(a)}=A(x_0a_0+x_1a_1+\dots+x_{n-1}a_{n-1})\\
&=x_0A(a_0)+x_1A(a_1)+\dots+x_{n-1}A(a_{n-1})\\
&=x_0(\mathcal{A}(a_0))_{(b)}+x_1(\mathcal{A}(a_1))_{(b)}+\dots+x_{n-1}(\mathcal{A}(a_{n-1}))_{(b)}.
\end{align}

Свака слика линеарниим пресликавањем је линеарна комбинација слика базе $(a).$ Због везе између линеарних пресликавања и матричног множења врло често се скраћује запис 
$$\mathcal{A}(v)\equiv \mathcal{A}v.$$

У случају линеарних оператора $\mathcal{A}:V\to V,$ њихове матрице обично посматрамо у односу на исту базу векторског простора $V,$ тј. $A=A\big(\mathcal{A},(a),(a)\big).$

**Пример 6.** Матрица идентичког оператора $\mathcal{I}:V\to V$ $n-$димензионалног векторског простора $V$ је јединична матрица у било којој бази овог векторског протсора
    $$I=\begin{bmatrix}1&0&\dots&0\\
    0&1&\dots&0\\
    \vdots&\vdots&\ddots&\vdots\\
    0&0&\dots&1\end{bmatrix}\in\mathcal{M}_{n\times n}.$$
Том приликом се и за домен и за кодомен користи иста база векторског простора.

Матрица нула-пресликавања $n-$димензионалног векторског простора у $m-$димензионални је $O=[0]\in\mathcal{M}_{m\times n}$ за сваки пар база ова два простора.

***

**Пример 7.** Преорука је да за вежбу покажете да је композиција линеарних пресликавања поново линеарно пресликавање. Операције сабирања функција и композиције линеарних пресликавања су у тесној вези са операцијама над матрицама. Наиме, ако су дате базе $(a),(b)$ и $(c)$ коначнодимензионалних векторских простора $V,U,W,$ редом, и у односу на њих матрице $A,B,C$ линеарних пресликавања $\mathcal{A}:V\to U,$ $\mathcal{B}:V\to U,$ $\mathcal{C}:U\to W,$ тада су познате и матрице линеарних пресликавања $\alpha\mathcal{A}+\beta\mathcal{B}:V\to U$ и $\mathcal{C}\circ\mathcal{A}:V\to W$ у односу на одговарајуће базе. Те матрице гласе $\alpha A+\beta B$ и $CA.$   
\begin{align}
&v\in V,\qquad v\equiv v_{(a)}=\begin{bmatrix} x_0&x_1&\dots&x_n\end{bmatrix}^T_{(a)},\\
&\mathcal{A}(v)\equiv (Av_{(a)})_{(b)},
    \qquad 
    \mathcal{B}(v)\equiv (Bv_{(a)})_{(b)},\\
    &(\alpha\mathcal{A}+\beta\mathcal{B})(v)=\alpha\mathcal{A}(v)+\beta\mathcal{B}(v)\equiv \alpha(Av_{(a)})_{(b)}+\beta(Bv_{(a)})_{(b)}
    =\big((\alpha A+\beta B)v_{(a)}\big)_{(b)},\\
    &(\mathcal{C}\circ\mathcal{A})(v)
    =\mathcal{C}\big(\mathcal{A}(v)\big) 
    \equiv \Big(C\big(Av_{(a)}\big)_{(b)}\Big)_{(c)}
    =\big(CAv_{(a)}\big)_{(c)}.
\end{align}

**Пример 8.** Нека је $\mathcal{A}:V\to U$ линеарно пресликавање између коначнодимензионалних векторских простора. Нека је $b_0,b_1,\dots,b_k$ база језгра $\mathcal{N}(A),$ и $b_{k+1},\dots,b_n$ допуна до базе простора домена $V.$ Матрица линеарног пресликавања $\mathcal{A}$ (у односу на произвољну базу простора $U$) тада има посебан блок облик
$$\big[O_{m\times(k+1)}\ |\ A_{m\times(n-k)}\big].$$

***

***
